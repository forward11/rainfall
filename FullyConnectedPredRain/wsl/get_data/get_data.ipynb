```python
import pandas as pd 
import numpy as np 
import os
```

## 筛选出乌鞘岭、武威和永昌

```python
dir = "C:\\Users\\11016\\Desktop\\data_0508\\2018-2020\\2019\\"
months = ['01','02','04','05','06','07','08','09','10','11','12']

for month in months:
    files = os.listdir(dir+month)
    wsl_temp, ww_temp, yc_temp = [],[],[]
    for file in files:
        file1 = dir+month+"\\"+file
        year,month,day,hour=file[:4],file[4:6],file[6:8],file[8:10]
        txt = np.loadtxt(file1)
        df = pd.DataFrame(txt)
        df1 = df[df[0]==52787] #wsl
        df1['year']=year
        df1['month']=month
        df1['day']=day
        df1['hour']=hour
        wsl_temp.append(df1)
        df2 = df[df[0]==52679] #ww
        df2['year']=year
        df2['month']=month
        df2['day']=day
        df2['hour']=hour
        ww_temp.append(df2)
        df3 = df[df[0]==52674] #yc
        df3['year']=year
        df3['month']=month
        df3['day']=day
        df3['hour']=hour
        yc_temp.append(df3)
    df_wsl = pd.concat(wsl_temp)
    df_ww = pd.concat(ww_temp)
    df_yc = pd.concat(yc_temp)
    df_wsl.to_csv(dir+"2019"+month+"_wsl.csv",index=False)
    df_ww.to_csv(dir+"2019"+month+"_ww.csv",index=False)
    df_yc.to_csv(dir+"2019"+month+"_yc.csv",index=False)
```
